```python
## Code for Final Exam Problem 3
## Serpentine formula plots


import math
import matplotlib.pyplot as plt


t=0
t11=[]
t22=[]
t33=[]
t44=[]
t55=[]
t66=[]
t77=[]
t88=[]
t99=[]
t1010=[]
time=[]



while t<10:
    t1=(-2*(math.pi/4)*math.sin(((t*math.pi*0.4)/1)+((t*2*math.pi*1)/10)))
    t2=(-2*(math.pi/4)*math.sin(((t*math.pi*0.4)/1)+((t*2*math.pi*2)/10)))
    t3=(-2*(math.pi/4)*math.sin(((t*math.pi*0.4)/1)+((t*2*math.pi*3)/10)))
    t4=(-2*(math.pi/4)*math.sin(((t*math.pi*0.4)/1)+((t*2*math.pi*4)/10)))
    t5=(-2*(math.pi/4)*math.sin(((t*math.pi*0.4)/1)+((t*2*math.pi*5)/10)))
    t6=(-2*(math.pi/4)*math.sin(((t*math.pi*0.4)/1)+((t*2*math.pi*6)/10)))
    t7=(-2*(math.pi/4)*math.sin(((t*math.pi*0.4)/1)+((t*2*math.pi*7)/10)))
    t8=(-2*(math.pi/4)*math.sin(((t*math.pi*0.4)/1)+((t*2*math.pi*8)/10)))
    t9=(-2*(math.pi/4)*math.sin(((t*math.pi*0.4)/1)+((t*2*math.pi*9)/10)))
    t10=(-2*(math.pi/4)*math.sin(((t*math.pi*0.4)/1)+((t*2*math.pi*10)/10)))
    t11.append(t1)
    t22.append(t2)
    t33.append(t3)
    t44.append(t4)
    t55.append(t5)
    t66.append(t6)
    t77.append(t7)
    t88.append(t8)
    t99.append(t9)
    t1010.append(t10)
    time.append(t)
    t=t+0.1
    
# Plot of Torque vs time
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6),(ax7, ax8), (ax9, ax10)) = plt.subplots(5, 2)
plt.subplots_adjust(wspace=0.4, hspace=1)
ax1.plot(time,t11)
ax1.set_title("t1 vs time")
ax2.plot(time,t22)
ax2.set_title("t2 vs time")
ax3.plot(time,t33)
ax3.set_title("t3 vs time")
ax4.plot(time,t44)
ax4.set_title("t4 vs time")
ax5.plot(time,t55)
ax5.set_title("t5 vs time")
ax6.plot(time,t66)
ax6.set_title("t6 vs time")
ax7.plot(time,t77)
ax7.set_title("t7 vs time")
ax8.plot(time,t88)
ax8.set_title("t8 vs time")
ax9.plot(time,t99)
ax9.set_title("t9 vs time")
ax10.plot(time,t1010)
ax10.set_title("t10 vs time")
```
